```python
from sklearn.model_selection import train_test_split
rate_train, rate_test = train_test_split(ratings.values, test_size=0.2, random_state=43)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
transformer = TfidfVectorizer()
tfidf = transformer.fit_transform(items['clean_description'])
```

```python
def get_items_rated_by_user(rate_matrix, user_id):
    """
    in each line of rate_matrix, we have infor: user_id, item_id, rating (scores), time_stamp
    we care about the first three values
    return (item_ids, scores) rated by user user_id
    """
    y = rate_matrix[:, 1] # all users
    # print(y)
    # item indices rated by user_id
    # we need to +1 to user_id since in the rate_matrix, id starts from 1 
    # while index in python starts from 0
    ids = np.where(y == user_id)[0] 
    # print(ids)
    item_ids = rate_matrix[ids, 2] # index starts from 0 
    # print(item_ids)
    scores = rate_matrix[ids, 0]
    # print(scores)
    return (item_ids, scores)
```

```python
from sklearn.linear_model import Ridge, RidgeCV, Lasso
from sklearn import linear_model
from sklearn.svm import SVR

d = tfidf.shape[1] # data dimension
W = np.zeros((d, n_users))
b = np.zeros((1, n_users))

for n in range(n_users):
    print(f"User {n}")  
    ids, scores = get_items_rated_by_user(rate_train, n)
    clf = Ridge(alpha=0.01, fit_intercept=True)
    # clf = RidgeCV(alphas=[1e-3, 1e-2, 1e-1, 1], cv=5)
    # clf = Lasso()
    # clf = linear_model.LinearRegression()
    # clf = SVR()
    Xhat = tfidf[ids, :]
    
    clf.fit(Xhat, scores) 
    W[:, n] = clf.coef_
    b[0, n] = clf.intercept_
```

```python
# predicted scores
Yhat = tfidf.dot(W) + b
```

```python
def recommend(Yhat, user_id, top, rates):
    a = np.zeros((n_items,))
    recommended_items = []
    items_rated_by_user, score = get_items_rated_by_user(rates, user_id)
    categories = set()
    for i, item in enumerate(items_rated_by_user):
        if score[i] >= 4:
            categories.add(str(items.loc[items['new_id']==item]['big_category'].values[0]))
        # category_index = items.loc[items['big_category'].isin(list(categories))].index
    # if not category_index:
# 
    for i in range(n_items):
        if i not in items_rated_by_user:
            if str(items.loc[items['new_id']==i]['big_category'].values[0]) in categories:
                a[i] = Yhat[i, user_id]
           
    if len(a) < top:
        recommended_items = np.argsort(a)[-len(a):]
    else:
        recommended_items = np.argsort(a)[-top:]
    return recommended_items
```

```python
from math import sqrt
def evaluate(Yhat, rates, W, b):
    se = 0
    cnt = 0
    for n in range(n_users):
        ids, scores_truth = get_items_rated_by_user(rates, n)
        scores_pred = Yhat[ids, n]
        e = scores_truth - scores_pred 
        se += (e*e).sum(axis = 0)
        cnt += e.size
    return sqrt(se/cnt)

print('RMSE for training:', evaluate(Yhat, rate_train, W, b))
print('RMSE for test    :', evaluate(Yhat, rate_test, W, b))
```

```python
def recommend(Yhat, user_id, top, rates):
    a = np.zeros((n_items,))
    recommended_items = []
    items_rated_by_user, score = get_items_rated_by_user(rates, user_id)
    for i in range(n_items):
        if i not in items_rated_by_user:
            a[i] = Yhat[i, user_id]
           
    if len(a) < top:
        recommended_items = np.argsort(a)[-len(a):]
    else:
        recommended_items = np.argsort(a)[-top:]
    return recommended_items
    
def evaluatePR(rate_test, top):
    sum_p = 0
    Pu = np.zeros((n_users,))
    for u in range(n_users):
        recommended_items = recommend(Yhat, u, top, rate_test)
        ids = np.where(rate_test[:, 1] == u)[0]
        rated_items = rate_test[ids, 1]
        for i in recommended_items:
            if i in rated_items:
                Pu[u] += 1
        sum_p += Pu[u]
    # so luong goi y tao ra = so user * top
    # so luong san pham chon boi nguoi dung = shape(rate_test)
    p = sum_p/(n_users * top)
    r = sum_p/(rate_test.shape[0] + 1)
#     print(sum_p)
    print('%d::%r::%r\r\n' % (top, p, r))
```
